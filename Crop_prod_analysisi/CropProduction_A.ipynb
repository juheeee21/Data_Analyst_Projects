## Total Production and Productivity

**Total production**

```python
df['Total Production'] = df['Production'] * df['Area']
```

```python
df['Total Production']
```

Remove Outliers

```python
Q1= df["Total Production"].quantile(0.40)
Q3 =df["Total Production"].quantile (0.60)
IQR = Q3 - Q1
df =df[(df["Total Production"] >= Q1-1.5*IQR) & (df ["Total Production"]<Q3)]
```

```python
Crop= df['Crop']
```

```python
def crop_category(crop):
    for i in ['Rice','Maize','Wheat','Barley','Varagu','Other Cereals & Millets']:
        if crop==i:
            return 'Cereal'
    for i in ['Moong','Urad','Arhar/Tur','Peas & beans','Masoor',
 'Other Kharif pulses','other misc. pulses','Ricebean (nagadal)'
 'Rajmash Kholar','Lentil','Samai','Blackgram','Korra','Cowpea(Lobia)',
 'Other Rabi pulses','Other Kharif pulses','Peas & beans(Pulses)'] :
        if crop==i:
            return 'Pulses'
    for i in ['Peach','Apple','Litchi','Pear','Plums','Ber','Sapota','Lemon'
 'Other Citrus Fruit','Water Melon','Jack Fruit','Grapes'
 'Pome Fruit','Citrus Fruit','Other Fresh Fruits','Mango']:
        if crop==i:
            return 'fruits'
    for i in ['Bean','Lab-Lab','Moth','Guar seed','Soyabean','Horse-gram']:
        if crop==i:
            return 'beans'
    for i in ['Turnip','Peas','Beet Root','Carrot','Yam','Ribed Guard',
 'Bitter Gourd','Cucumber','Drum Stick','Cauliflower','Beans',
 'Bhindi','Tomato','Brinjal','Khesari','Sweet potato','Potato']:
        if crop==i:
            return 'vegetables'
    for i in ['Perilla','Ginger','Cardamom','Black pepper','Dry ginger']:
        if crop==i:
             return 'spices'
    for i in ['other fibres','Kapas','Jute & mesta','Jute','Mesta','Cotton']:
        if crop==i:
            return 'fibres'
    for i in ['Arcanut (Processed)','Atcanut (Raw)','Cashewnut Processed']:
        if crop==i:
             return 'Nuts'
    for i in ['other oilseeds','Safflower','Niger seed','Castor seed','Oilseeds total']:
        if crop==i:
            return 'Oilseeds'
    for i in ['Tobacco','Coffee','Tea','Sugarcane','Rubber']:
        if crop==i:
            return 'Commercial'

df['crop_category']=df['Crop'].apply(crop_category)

```
